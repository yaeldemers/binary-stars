```python
#Function to generate the light curve given the data from NASA's Transiting Exoplanet Survey Satellite (TESS)
def generating_light_curve(data, title):
    
    #generating array of observation dates (x-axis)
    date = np.transpose(data)[0]
    date = np.array(date)
    time = date.astype(np.float64)
    
    #generating array of magnitude recorded (y-axis)
    magnitude = np.transpose(data)[1]
    magnitude = np.array(magnitude)
    TESS_magnitude = magnitude.astype(np.float64)

    #creating plot
    F1 = plt.figure(figsize = (10,7))
    plt.scatter(time, TESS_magnitude, s=15)
    
    #adding periodicity text box
    plt.text(0.58, 0.2, r'P = 2.867d',
             verticalalignment='top', horizontalalignment='right',
             transform=plt.gca().transAxes,
             color='black', fontsize=15, bbox=dict(facecolor='white', 
                                                   edgecolor='black', alpha=0.8))

    #Plot Specifications
    plt.xlabel("Days", fontsize = 20)
    plt.ylabel('TESS Magnitude', fontsize = 20)
    plt.tick_params(bottom=True, top=True, left=True, right=True, which='major', 
                direction = "out", labelsize=15, width=1.3, color='black')
    plt.tick_params(top= True, right = True, axis='both', which='minor', direction='out', 
                labelsize = 12, length=5, width=1.3)
    plt.gca().invert_yaxis() # invert y-axis
    plt.title(title, fontsize = 25)
    plt.tight_layout()
        
    plt.show()
    
    F1.savefig("TESS Light curve", format="png", dpi=1200)
```

```python
#Reading .csv
file = open("Algol 2.csv")
csvreader = csv.reader(file)
algol2 = []
for row in csvreader:
    algol2.append(row)
```

```python
generating_light_curve(algol2, r"Algol ($\beta$ Persei) Light Curve")
```
